```python
import pandas as pd
```

```python


file_path1 = 'Uncleaned datasets/Environment_Temperature_change_E_All_Data_NOFLAG.csv'
data1 = pd.read_csv(file_path1, encoding='unicode_escape')
dataframe1 = pd.DataFrame(data1)
dataframe1.head()
```

```python
file_path2 = 'Uncleaned datasets/co2_emissions_kt_by_country.csv'
data2 = pd.read_csv(file_path2, encoding='unicode_escape')
dataframe2 = pd.DataFrame(data2)
dataframe2.head()
```

```python
file_path3 = 'Uncleaned datasets/ElNino.csv'
data3 = pd.read_csv(file_path3, encoding='utf-8')
dataframe3 = pd.DataFrame(data3)
dataframe3.head()
```

```python
file_path4 = 'Uncleaned datasets/methane.csv'
data4 = pd.read_csv(file_path4, encoding='unicode_escape')
dataframe4 = pd.DataFrame(data4)
dataframe4.head()
```

```python
# Remove all rows with empty values
df1_cleaned = dataframe1.dropna()
df2_cleaned = dataframe2.dropna()
df3_cleaned = dataframe3.dropna()
df4_cleaned = dataframe4.dropna()
```

```python
# Rename the columns to be more understandable
df3_cleaned = df3_cleaned.rename(columns={"JJA": "Jan",
                             "JAS": "Feb",
                             "ASO": "Mar",
                             "SON": "Apr",
                             "OND": "May",
                             "NDJ": "Jun",
                             "DJF": "Jul",
                             "JFM": "Aug",
                             "FMA": "Sep",
                             "MAM": "Oct",
                             "AMJ": "Nov",
                             "MJJ": "Dec",})
```

```python
# Remove unnecessary column
df4_cleaned = df4_cleaned.drop("Unnamed: 0", axis = 1)
```

```python
# Change the years from year-year to just the one left of the -, so 1950-1951 becomes 1950.
df3_cleaned['Season'] = df3_cleaned['Season'].str.split('-').str[0]
df3_cleaned = df3_cleaned.rename(columns = {'Season':'year'})
```

```python
df1_cleaned.to_csv('Temperature_change_cleaned.csv', index = False)
df2_cleaned.to_csv('CO2_emissions_cleaned.csv', index = False)
df3_cleaned.to_csv('ElNino_LaNina_cleaned.csv', index = False)
df4_cleaned.to_csv('Methane_emissions_cleaned.csv', index = False)
```

```python
long_lat_data = pd.read_csv('world_country_and_usa_states_latitude_and_longitude_values.csv')
long_lat_dataframe = pd.DataFrame(long_lat_data)
long_lat_dataframe = long_lat_dataframe.drop(['country_code', 'usa_state_code', 'usa_state_latitude', 'usa_state_longitude', 'usa_state'], axis=1)

long_lat_dataframe.to_csv('new_long_lat.csv')

print(long_lat_dataframe)
```

```python
co2_df = pd.DataFrame(pd.read_csv('new_co2.csv'))
methane_df = pd.DataFrame(pd.read_csv('new_methane.csv'))
temperature_df = pd.DataFrame(pd.read_csv('new_temperature.csv'))
elnino_df = pd.DataFrame(pd.read_csv('new_elnino_lanina.csv'))
long_let_df = pd.DataFrame(pd.read_csv('new_long_lat.csv'))

co2_df = co2_df.rename(columns={'country_name':'Country'})
co2_df = co2_df.rename(columns={'year':'Year'})

all_data = pd.merge(co2_df, methane_df, on=['Country', 'Year'], how='inner')
all_data = all_data.drop(['country_code'], axis=1)

all_data = all_data.rename(columns={'z-score_x':'z-score-co2'})
all_data = all_data.rename(columns={'z-score_y':'z-score-methane'})
all_data = all_data.drop(['value', 'Methane value'], axis=1)

only_meteorogical_df = temperature_df[temperature_df['Months'] == 'Meteorological year']
only_meteorogical_df = only_meteorogical_df.rename(columns={'Area':'Country'})

all_data = pd.merge(all_data, only_meteorogical_df, on=['Country', 'Year'], how='inner')
all_data = all_data.drop(['Area Code', 'Months Code', 'Months', 'Element Code', 'Element', 'Unit', 'z-score'], axis=1)

elnino_df = elnino_df.rename(columns={'year':'Year'})
all_data = pd.merge(all_data, elnino_df, on=['Year'], how='inner')
all_data = all_data.drop(['Unnamed: 0'], axis=1)
all_data = all_data.rename(columns={'z-score':'z-score-elnino-lanina'})

long_let_df = long_let_df.rename(columns={'country':'Country'})
all_data = pd.merge(all_data, long_let_df, on=['Country'], how='inner')
all_data = all_data.drop(['Unnamed: 0'], axis=1)

all_data.to_csv('all_data_and_long_lat.csv', index=False)

print(all_data)
```
